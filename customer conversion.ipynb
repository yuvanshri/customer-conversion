```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_data is already defined and loaded
df_data = pd.read_csv('/content/test_data.csv')

# Select only the numerical columns for the mean calculation
numerical_cols_for_mean = df_data.select_dtypes(include=np.number).columns.tolist()

# Exclude the grouping column 'price' if it's in the numerical columns list
if 'price' in numerical_cols_for_mean:
    numerical_cols_for_mean.remove('price')

# Calculate the mean only on numerical columns after grouping by 'price'
# Ensure that there are numerical columns to calculate the mean on
if numerical_cols_for_mean:
    grouped_status = df_data.groupby('price')[numerical_cols_for_mean].mean().T
    # Check if grouped_status is not empty before plotting
    if not grouped_status.empty:
        grouped_status.plot(kind='bar', figsize=(15, 6))
        plt.title('Mean of Numerical Features by Price') # Add a title for clarity
        plt.xlabel('Feature') # Add an x-axis label
        plt.ylabel('Mean Value') # Add a y-axis label
        plt.show()
    else:
        print("No numerical columns available to calculate the mean for plotting.")
else:
    print("No numerical columns found in the DataFrame to calculate the mean.")

# Rest of your plotting code (already in the notebook) can follow here if needed,
# using the original numerical_cols if you want to plot all numerical columns.

# Example of how to redefine numerical_cols if you need it for other plots
numerical_cols = df_data.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(10, len(numerical_cols) * 2))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(len(numerical_cols) // 2 + 1, 2, i)
    sns.boxplot(x=df_data[col], color='pink')
    plt.title(col)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, len(numerical_cols) * 2))
for i, col in enumerate(numerical_cols, 1):
  plt.subplot(len(numerical_cols) // 2 + 1, 2, i)
  sns.histplot(x=df_data[col], bins=20, color='blue', kde=True)
  plt.title(col)
plt.tight_layout()
plt.show()

skewed_features = df_data.drop('price', axis=1).columns
print(skewed_features)

df_data.info()

num_cols = df_data.select_dtypes(include=np.number).columns.tolist()

# Iterate through each numerical column to generate individual plots
for col in num_cols:
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df_data[col], kde=True, bins=30, color='skyblue')
    plt.title(f'distribution of {col}')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_data[col], color='orange')
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()

df_data.columns


df_data.columns

plt.figure(figsize=(12, 4))
sns.heatmap(df_data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f') # Added numeric_only=True
plt.title('correlation heatmap')
plt.show()
```
